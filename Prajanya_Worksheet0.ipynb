```python
#Task1
def convert_units(value, conversion_type):
    """
    Converts a numeric value between supported measurement units.

    Parameters:
        value (float): The number to be converted.
        conversion_type (str): Type of conversion. Supported values:
            '1' - Meters to Feet
            '2' - Feet to Meters
            '3' - Kilograms to Pounds
            '4' - Pounds to Kilograms

    Returns:
        float: Converted value if valid conversion type.
        None: If conversion type is invalid.
    """

    if conversion_type == '1':
        return value * 3.28084
    elif conversion_type == '2':
        return value / 3.28084
    elif conversion_type == '3':
        return value * 2.20462
    elif conversion_type == '4':
        return value / 2.20462
    else:
        return None


print("=== Unit Converter ===")

while True:
    print("\nSelect Conversion Type:")
    print("1) Meters to Feet")
    print("2) Feet to Meters")
    print("3) Kilograms to Pounds")
    print("4) Pounds to Kilograms")
    print("5) Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == '5':
        print("Goodbye!")
        break

    try:
        value = float(input("Enter the value to convert: "))
    except ValueError:
        print("Error: Please enter a valid numeric value.")
        continue

    result = convert_units(value, choice)

    if result is None:
        print("Invalid conversion choice. Please try again.")
    else:
        print(f"Converted value: {result:.4f}")
```

```python
#Task2
def compute_sum(numbers):
    """
    Calculate the total sum of a list of numbers.

    Parameters:
        numbers (list of float): A non-empty list of numeric values.

    Returns:
        float: The sum of all elements in the list.
    """
    return sum(numbers)


def compute_average(numbers):
    """
    Calculate the arithmetic mean (average) of a list of numbers.

    Parameters:
        numbers (list of float): A non-empty list of numeric values.

    Returns:
        float: The average value of the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot compute average of an empty list.")
    return sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Find the largest number in a list.

    Parameters:
        numbers (list of float): A non-empty list of numeric values.

    Returns:
        float: The maximum value in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot find maximum of an empty list.")
    return max(numbers)


def compute_minimum(numbers):
    """
    Find the smallest number in a list.

    Parameters:
        numbers (list of float): A non-empty list of numeric values.

    Returns:
        float: The minimum value in the list.

    Raises:
        ValueError: If the list is empty.
    """
    if not numbers:
        raise ValueError("Cannot find minimum of an empty list.")
    return min(numbers)


def parse_numbers(user_input):
    """
    Convert a space-separated string into a list of floats.

    Parameters:
        user_input (str): A string containing numbers separated by spaces.

    Returns:
        list of float: List of converted numeric values.

    Raises:
        ValueError: If input is empty or contains non-numeric values.
    """
    tokens = user_input.strip().split()

    if not tokens:
        raise ValueError("No numbers entered.")

    return [float(token) for token in tokens]


# ================= MAIN PROGRAM =================

print("=== Mathematical Operations on a List ===")

operations = {
    "1": (compute_sum, "Sum"),
    "2": (compute_average, "Average"),
    "3": (compute_maximum, "Maximum"),
    "4": (compute_minimum, "Minimum"),
}

while True:
    print("\nChoose an operation:")
    print("1) Sum")
    print("2) Average")
    print("3) Maximum")
    print("4) Minimum")
    print("5) Quit")

    choice = input("Enter your choice (1-5): ").strip()

    if choice == "5":
        print("Goodbye!")
        break

    if choice not in operations:
        print("Invalid choice. Please select 1-5.")
        continue

    user_input = input("Enter numbers separated by spaces: ")

    try:
        numbers = parse_numbers(user_input)
        function, name = operations[choice]
        result = function(numbers)
    except ValueError as e:
        print(f"Error: {e}")
        continue

    print(f"\nNumbers: {numbers}")
    print(f"{name}: {result}")
```

```python
#4.2.1
def extract_every_other(lst):
    """
    Extract every other element from a list, starting from the first element.

    Parameters:
        lst (list): A list of elements.

    Returns:
        list: A new list containing every other element from the original list.
    """
    return lst[::2]



numbers = [1, 2, 3, 4, 5, 6]
result = extract_every_other(numbers)

print("Original list:", numbers)
print("Every other element:", result)
```

```python
#4.2.2
def get_sublist(lst, start, end):
    """
    Return a sublist from a given list starting at index 'start'
    and ending at index 'end' (inclusive).

    Parameters:
        lst (list): The original list.
        start (int): Starting index.
        end (int): Ending index (inclusive).

    Returns:
        list: A new list containing elements from index start to end.

    Raises:
        ValueError: If start or end indices are invalid.
    """
    if start < 0 or end >= len(lst) or start > end:
        raise ValueError("Invalid start or end index.")

    return lst[start:end + 1]


# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = get_sublist(numbers, 2, 4)

print("Original list:", numbers)
print("Sublist:", result)
```

```python
#4.2.3
def reverse_list(lst):
    """
    Reverse a list using slicing.

    Parameters:
        lst (list): The original list.

    Returns:
        list: A new list containing the elements of lst in reverse order.
    """
    return lst[::-1]


# Example usage
numbers = [1, 2, 3, 4, 5]
result = reverse_list(numbers)

print("Original list:", numbers)
print("Reversed list:", result)
```

```python
#4.2.4
def remove_first_last(lst):
    """
    Remove the first and last elements of a list using slicing.

    Parameters:
        lst (list): The original list.

    Returns:
        list: A new list without the first and last elements.
    """
    return lst[1:-1]


# Example
print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
#4.2.5
def get_first_n(lst, n):
    """
    Extract the first n elements from a list using slicing.

    Parameters:
        lst (list): The original list.
        n (int): Number of elements to extract.

    Returns:
        list: A list containing the first n elements.
    """
    return lst[:n]


# Example
print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
#4.2.6
def get_last_n(lst, n):
    """
    Extract the last n elements from a list using slicing.

    Parameters:
        lst (list): The original list.
        n (int): Number of elements to extract from the end.

    Returns:
        list: A list containing the last n elements.
    """
    return lst[-n:]


# Example
print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
#4.2.7
def reverse_skip(lst):
    """
    Extract every second element in reverse order starting
    from the second-to-last element.

    Parameters:
        lst (list): The original list.

    Returns:
        list: A new list containing every second element
              from the second-to-last, moving backward.
    """
    return lst[-2::-2]


# Example
print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

```python
#4.3.1
def flatten(lst):
    """
    Flatten a nested list (one level deep) into a single list.

    Parameters:
        lst (list): A list containing sublists.

    Returns:
        list: A flattened list.
    """
    result = []
    for sublist in lst:
        result.extend(sublist)
    return result


# Example
print(flatten([[1, 2], [3, 4], [5]]))
```

```python
#4.3.2
def access_nested_element(lst, indices):
    """
    Access a specific element in a nested list using a list of indices.

    Parameters:
        lst (list): Nested list.
        indices (list): List of indices representing the path.

    Returns:
        Element at the specified position.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element


# Example
lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst, [1, 2]))
```

```python
#4.3.3
def sum_nested(lst):
    """
    Calculate the sum of all elements in a nested list (any depth).

    Parameters:
        lst (list): Nested list containing numbers.

    Returns:
        int/float: Sum of all numbers.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# Example
print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
#4.3.4
def remove_element(lst, elem):
    """
    Remove all occurrences of elem from a nested list.

    Parameters:
        lst (list): Nested list.
        elem: Element to remove.

    Returns:
        list: Modified nested list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            cleaned = remove_element(item, elem)
            result.append(cleaned)
        elif item != elem:
            result.append(item)
    return result


# Example
print(remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

```python
#4.3.5
def find_max(lst):
    """
    Find the maximum element in a nested list.

    Parameters:
        lst (list): Nested list of numbers.

    Returns:
        int/float: Maximum value.
    """
    max_value = float("-inf")
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)
    return max_value


# Example
print(find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
#4.3.6
def count_occurrences(lst, elem):
    """
    Count occurrences of elem in a nested list.

    Parameters:
        lst (list): Nested list.
        elem: Element to count.

    Returns:
        int: Number of occurrences.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# Example
print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

```python
#4.3.7
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters:
        lst (list): Deeply nested list.

    Returns:
        list: Completely flattened list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# Example
print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

```python
#4.3.8
def average_nested(lst):
    """
    Calculate the average of all elements in a nested list.

    Parameters:
        lst (list): Nested list of numbers.

    Returns:
        float: Average of all elements.
    """
    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list)


# Example
print(average_nested([[1, 2], [3, 4], [5, 6]]))
```
